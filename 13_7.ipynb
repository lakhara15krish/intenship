```python

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python

# Load CSV file
df = pd.read_csv("/content/energy_data.csv")

# Display first 5 rows
df.head()
```

```python
# Dataset Information
print(df.info())
```

```python
# Check Shape
print("Rows and Columns:", df.shape)
```

```python

# Check Data Types
print(df.dtypes)
```

```python

# Remove Missing Values (If Any)
df = df.dropna()

print(df.isnull().sum())
```

```python
# Dataset Summary
print(df.describe())
```

```python
# Convert Timestamp into DateTime
df["timestamp"] = pd.to_datetime(df["timestamp"], dayfirst=True)

df.head()
```

```python

# Feature Engineering
df["Hour"] = df["timestamp"].dt.hour
df["Day"] = df["timestamp"].dt.day
df["Month"] = df["timestamp"].dt.month

df.head()
```

```python
# Encode Location Column
encoder = LabelEncoder()

df["location_encoded"] = encoder.fit_transform(df["location"])

df.head()
```

```python
# Correlation Matrix
plt.figure(figsize=(8,6))

sns.heatmap(
    df[["energy_consumption","temperature","Hour","location_encoded"]].corr(),
    annot=True,
    cmap="Reds"
)

plt.title("Correlation Matrix")
plt.show()
```

```python
# Calculate mean and median
mean_energy = df["energy_consumption"].mean()
median_energy = df["energy_consumption"].median()

# Distribution of Energy Consumption (Enhanced)
plt.figure(figsize=(10, 6))

sns.histplot(df["energy_consumption"], bins=40, kde=True, color='skyblue', edgecolor='black')

# Add mean and median lines
plt.axvline(mean_energy, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_energy:.2f}')
plt.axvline(median_energy, color='green', linestyle='-', linewidth=2, label=f'Median: {median_energy:.2f}')

plt.title("Enhanced Energy Consumption Distribution")
plt.xlabel("Energy Consumption")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Temperature Distribution
plt.figure(figsize=(8,5))

sns.histplot(df["temperature"], bins=30, color="orange", kde=True)

plt.title("Temperature Distribution")
plt.show()
```

```python
# Energy Consumption by Location
plt.figure(figsize=(8,5))

sns.boxplot(
    x="location",
    y="energy_consumption",
    data=df,
    color='red', # Set the main color to red
    boxprops=dict(alpha=0.7) # Add transparency for a 'shady' effect
)

plt.title("Energy Consumption by Location")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10,7))

sns.scatterplot(
    x="temperature",
    y="energy_consumption",
    hue="location",
    data=df,
    palette='viridis', # Change color palette
    s=100, # Increase marker size
    alpha=0.7, # Add transparency
    edgecolor='w' # Add white edge to markers
)

plt.title("Temperature vs Energy Consumption by Location", fontsize=14)
plt.xlabel("Temperature", fontsize=12)
plt.ylabel("Energy Consumption", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Prepare Features for K-Means
X = df[
    [
        "energy_consumption",
        "temperature",
        "Hour",
        "location_encoded"
    ]
]
```

```python
# Standardize Data
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
# Elbow Method
wcss = []

for i in range(1,11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X_scaled)

    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,6)) # Increase figure size for better visibility

plt.plot(range(1,11), wcss, marker='o', linestyle='--', color='purple', markersize=8, markerfacecolor='gold') # Enhanced line and marker style

plt.xlabel("Number of Clusters (K)", fontsize=12) # Larger font size for x-label
plt.ylabel("Within-Cluster Sum of Squares (WCSS)", fontsize=12) # Larger font size for y-label

plt.title("Elbow Method for Optimal K", fontsize=14) # More descriptive title with larger font
plt.grid(True, linestyle=':', alpha=0.7) # Refined grid style
plt.xticks(range(1,11)) # Ensure all k values are shown on x-axis
plt.show()
```

```python

# Apply K-Means
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(X_scaled)

df["Cluster"] = clusters

df.head()
```

```python
# Cluster Visualization
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="temperature",
    y="energy_consumption",
    hue="Cluster",
    palette="Set1",
    data=df
)

plt.title("K-Means Clusters")

plt.show()
```

```python

# Cluster Count
print(df["Cluster"].value_counts())
```

```python

# Pair Plot
sns.pairplot(
    df,
    vars=[
        "energy_consumption",
        "temperature",
        "Hour"
    ],
    hue="Cluster"
)

plt.show()
```
